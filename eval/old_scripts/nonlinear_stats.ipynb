```python
import numpy as np
subjects = !ls '/scr/ilz1/nonlinear_registration/results_all_subjects/'
```

```python
nmi_diff_array=np.array([])
cr_diff_array=np.array([])
mask_diff_array=np.array([])

nmi_array= np.empty(shape=(1,2))
cr_array=np.empty(shape=(1,2))
mask_array=np.empty(shape=(1,2))

for subject in range(len(subjects)):
    path = '/scr/ilz1/nonlinear_registration/results_all_subjects/%s/eval/txt/%s/metrics.txt' %(subjects[subject],subjects[subject])
    metrics_array=np.loadtxt(path, dtype='float', delimiter=' ')
    nmi_difference=metrics_array[1,1]-metrics_array[1,0]
    cr_difference =metrics_array[3,1]-metrics_array[3,0]
    mask_difference=metrics_array[5,1]-metrics_array[5,0]
    
    nmi_diff_array = np.append(nmi_diff_array,nmi_difference)
    cr_diff_array = np.append(cr_diff_array,cr_difference)
    mask_diff_array = np.append(mask_diff_array,mask_difference)
    
    nmi = metrics_array[1,:]
    cr = metrics_array[3,:]
    mask = metrics_array[5,:]
    nmi_array = np.concatenate((nmi_array,nmi.reshape(1,np.size(nmi))),axis=0)
    cr_array = np.concatenate((cr_array,cr.reshape(1,np.size(cr))),axis=0)
    mask_array = np.concatenate((mask_array,mask.reshape(1,np.size(mask))),axis=0)
```

```python
from scipy import stats as st
nmi_test=st.ttest_1samp(nmi_diff_array, 0)
cr_test=st.ttest_1samp(cr_diff_array, 0)
mask_test=st.ttest_1samp(mask_diff_array, 0)

print nmi_test
print cr_test
print mask_test
```
