## Robust min and max of fields across subjects

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy as sp
import os
```

# CHECK BEFORE RUNNING

```python
version='version_11'
all_versions=['8','10','5', '6','11']
```

```python
subjects=[]
f = open('/scr/jessica2/Schaare/LEMON/done_freesurfer.txt','r')
for line in f:
    subjects.append(line.strip())
```

```python
methods = ['nonlin', 'fmap','topup']
```

```python
file_path='/scr/kansas1/huntenburg/eval/'+version+'/%s/fieldcompare/min_max.txt'
save_path='/scr/kansas1/huntenburg/eval/'+version+'/figures/group/'

if not os.path.isdir('/scr/kansas1/huntenburg/eval/'+version+'/figures/'):
    os.mkdir('/scr/kansas1/huntenburg/eval/'+version+'/figures/')
```

### Reading values into arrays

```python
for method in methods:
    if method=='nonlin': 
        row=0
    if method=='fmap': 
        row=1
    if method=='topup': 
        row=2
            
    # initiate empty array and min/max values
    min_array=np.array([])
    max_array=np.array([])
    diff_array=np.array([])
    min_min=[0, 'none']
    max_max=[0,'none']
    min_diff=[100,'none']
    max_diff=[0,'none']
        
    # loop over subjects, read in metrics file
    for subject in range(len(subjects)):
        path = file_path %(subjects[subject])
        stats_array=np.loadtxt(path, dtype='float', delimiter=' ')
        
        # define values
        min_val=stats_array[row, 0]
        max_val=stats_array[row, 1]
        diff_val=abs(min_val)+abs(max_val)
        
        # compare to min/max values and update if applicable
        if min_val < min_min[0]:
            min_min=[min_val,subjects[subject]]
        if max_val > max_max[0]:
            max_max=[max_val, subjects[subject]]
        if diff_val < min_diff[0]:
            min_diff=[diff_val, subjects[subject]]
        if diff_val > max_diff[0]:
            max_diff=[diff_val, subjects[subject]]
            
        # append values to arrays
        min_array = np.append(min_array,min_val)
        max_array = np.append(max_array,max_val)
        diff_array = np.append(diff_array,diff_val)
        
    # save array of all subjects to file
    if not os.path.isdir(save_path):
        os.mkdir(save_path)
    
    if not os.path.isdir(save_path+'arrays/'):
        os.mkdir(save_path+'arrays/')
    np.save(save_path+'arrays/'+method+'_field_min.npy', min_array)
    np.save(save_path+'arrays/'+method+'_field_max.npy', max_array)
    np.save(save_path+'arrays/'+method+'_field_diff.npy', diff_array)
    
    ## one sample t test if improvement differs from zero
    #[ttest_t, ttest_p]=sp.stats.ttest_1samp(diff_array, 0)
    #
    ## save min/max values and ttest to text
    #stats_array=np.array([['ttest improvement to linear[t,p] ',str([ttest_t, ttest_p])]])
    #stats_array=np.append(stats_array, [['least improved to linear[val,subj] ',str(diff_min)]], axis=0)
    #stats_array=np.append(stats_array, [['most improved to linear[val,subj] ',str(diff_max)]], axis=0)
    #stats_array=np.append(stats_array, [['min similarity[val,subj] ',str(abs_min)]], axis=0)
    #stats_array=np.append(stats_array, [['max similarity[val,subj] ',str(abs_max)]], axis=0)

    #if not os.path.isdir(save_path+'stats/'):
    #    os.mkdir(save_path+'stats/')
    #np.savetxt(save_path+'stats/'+method+'_'+metric+'_stats.txt', stats_array, delimiter=' ', fmt='%s')
    
    # print stats for quick check
    #print method+' minimum '+metric+' similarity = '+str(abs_min)
    #print method+' maximum '+metric+' similarity = '+str(abs_max)
    #print method+' least '+metric+' improvement to linear = '+str(diff_min)
    #print method+' highest '+metric+' improvement to linear = '+str(diff_max)
    #print method+' '+metric+' improvement diff from zero [t,p] = '+str([ttest_t, ttest_p])
    #if ttest_p <= 0.01:
    #    if ttest_t >=0:
    #        print metric+' '+method+' significant improvement[t,p] = '+str([ttest_t, ttest_p])
    #    elif ttest_t <=0:
    #        print metric+' '+method+' significant deterioration[t,p] = '+str([ttest_t, ttest_p])
    
    
print min_min
print max_max
print min_diff
print max_diff
```

```python
fig = plt.figure(figsize=(35,5))

for i  in ['min', 'max', 'diff']:
    fmap=np.load(save_path+'arrays/fmap_field_'+i+'.npy')
    topup=np.load(save_path+'arrays/topup_field_'+i+'.npy')

    nonlin_names=[]
    all_nonlin=np.zeros((len(fmap),len(all_versions)))
    
    row=0
    for v in all_versions:
        nonlin=np.load('/scr/kansas1/huntenburg/eval/version_'+v+'/figures/group/arrays/nonlin_field_'+i+'.npy')
        nonlin_names=nonlin_names+['nonlinear '+v]
        all_nonlin[:,row]=nonlin
        row+=1
    
    #plot_list=[fmap]+[topup]
    plot_list=[fmap]
    for j in range(len(all_versions)):
        plot_list=plot_list+[all_nonlin[:,j]]
    
    #names=['fieldmap']+['topup']+nonlin_names
    names=['fieldmap']+nonlin_names
        
        
    if i == 'min':
        ax = fig.add_subplot(131)
    elif i == 'max':
        ax = fig.add_subplot(132)
    elif i =='diff':
        ax = fig.add_subplot(133)
    
    sns.violinplot(plot_list,
            names=names,
            inner="points", 
            color=pal,
            #alpha=.6
            )
    sns.axlabel('',i)

plt.savefig(save_path+'group_field_minmaxdiff_violins.svg')
```
