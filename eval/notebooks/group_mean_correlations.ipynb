## Correlations of mean epi images across subjects 

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy as sp
import os
```

# CHECK BEFORE RUNNING

```python
version = 'version_8'
all_versions=['8','5','11']
```

```python
subjects=[]
f = open('/scr/jessica2/Schaare/LEMON/done_freesurfer.txt','r')
for line in f:
    subjects.append(line.strip())
```

```python
spaces=['mni','mni_ofc','indv']
correlations = ['lin_nonlin', 'lin_fmap', 'lin_topup', 'nonlin_fmap', 'nonlin_topup', 'topup_fmap']
```

```python
save_path='/scr/kansas1/huntenburg/eval/'+version+'/figures/group/'

if not os.path.isdir('/scr/kansas1/huntenburg/eval/'+version+'/figures/'):
    os.mkdir('/scr/kansas1/huntenburg/eval/'+version+'/figures/')
```

### Reading values into arrays

```python
for space in spaces:
    if space == 'mni':
        file_path='/scr/kansas1/huntenburg/eval/'+version+'/%s/correlation/mni/mean_correlation_mni.txt'
    elif space == 'mni_ofc':
        file_path='/scr/kansas1/huntenburg/eval/'+version+'/%s/correlation/mni_ofc/mean_correlation_mni_ofc.txt'
    elif space == 'indv':
        file_path='/scr/kansas1/huntenburg/eval/'+version+'/%s/correlation/indv/mean_correlation_indv.txt'


    for correlation in correlations:
        if correlation=='lin_nonlin': 
            row=0
        if correlation=='lin_fmap': 
            row=1
        if correlation=='lin_topup': 
            row=2
        if correlation=='nonlin_fmap': 
            row=3
        if correlation=='nonlin_topup': 
            row=4
        if correlation=='topup_fmap': 
            row=5

        # initiate empty array and min/max values
        r_array=np.array([])
        r_min=[1, 'none']
        r_max=[0,'none']

        # loop over subjects, read in metrics file
        for subject in range(len(subjects)):
            path = file_path %(subjects[subject])
            stats_array=np.loadtxt(path, dtype='float', delimiter=' ')

            # define abs and diff values
            r_val=stats_array[row, 0]

            # compare to min/max values and update if applicable
            if r_val < r_min[0]:
                r_min=[r_val,subjects[subject]]
            if r_val > r_max[0]:
                r_max=[r_val, subjects[subject]]

            # append values to arrays
            r_array = np.append(r_array,r_val)

        # save array of all subjects to file
        if not os.path.isdir(save_path):
            os.mkdir(save_path)

        if not os.path.isdir(save_path+'arrays/'):
            os.mkdir(save_path+'arrays/')
        np.save(save_path+'arrays/'+space+'_'+correlation+'_corr.npy', r_array)
    
    ## one sample t test if improvement differs from zero
    #[ttest_t, ttest_p]=sp.stats.ttest_1samp(diff_array, 0)
    #
    ## save min/max values and ttest to text
    #stats_array=np.array([['ttest improvement to linear[t,p] ',str([ttest_t, ttest_p])]])
    #stats_array=np.append(stats_array, [['least improved to linear[val,subj] ',str(diff_min)]], axis=0)
    #stats_array=np.append(stats_array, [['most improved to linear[val,subj] ',str(diff_max)]], axis=0)
    #stats_array=np.append(stats_array, [['min similarity[val,subj] ',str(abs_min)]], axis=0)
    #stats_array=np.append(stats_array, [['max similarity[val,subj] ',str(abs_max)]], axis=0)

    #if not os.path.isdir(save_path+'stats/'):
    #    os.mkdir(save_path+'stats/')
    #np.savetxt(save_path+'stats/'+method+'_'+metric+'_stats.txt', stats_array, delimiter=' ', fmt='%s')
    
    # print stats for quick check
    #print method+' minimum '+metric+' similarity = '+str(abs_min)
    #print method+' maximum '+metric+' similarity = '+str(abs_max)
    #print method+' least '+metric+' improvement to linear = '+str(diff_min)
    #print method+' highest '+metric+' improvement to linear = '+str(diff_max)
    #print method+' '+metric+' improvement diff from zero [t,p] = '+str([ttest_t, ttest_p])
    #if ttest_p <= 0.01:
    #    if ttest_t >=0:
    #        print metric+' '+method+' significant improvement[t,p] = '+str([ttest_t, ttest_p])
    #    elif ttest_t <=0:
    #        print metric+' '+method+' significant deterioration[t,p] = '+str([ttest_t, ttest_p])
```

```python
fig = plt.figure(figsize=(20,20))

for space in spaces:

    
    lin_fmap_r=np.load(save_path+'arrays/'+space+'_lin_fmap_corr.npy')
    lin_topup_r=np.load(save_path+'arrays/'+space+'_lin_topup_corr.npy')
    topup_fmap_r=np.load(save_path+'arrays/'+space+'_topup_fmap_corr.npy')
    
    all_lin_nonlin_r=np.zeros((len(lin_fmap_r),len(all_versions)))
    all_nonlin_fmap_r=np.zeros((len(lin_fmap_r),len(all_versions)))
    all_nonlin_topup_r=np.zeros((len(lin_fmap_r),len(all_versions)))
    
    fmap_names=[]
    topup_names=[]
    
    
    row=0
    for v in all_versions:
        lin_nonlin_r=np.load('/scr/kansas1/huntenburg/eval/version_'+v+'/figures/group/arrays/'+space+'_lin_nonlin_corr.npy')
        nonlin_fmap_r=np.load('/scr/kansas1/huntenburg/eval/version_'+v+'/figures/group/arrays/'+space+'_nonlin_fmap_corr.npy')
        nonlin_topup_r=np.load('/scr/kansas1/huntenburg/eval/version_'+v+'/figures/group/arrays/'+space+'_nonlin_topup_corr.npy')
    
        all_lin_nonlin_r[:,row]=lin_nonlin_r
        all_nonlin_fmap_r[:,row]=nonlin_fmap_r
        all_nonlin_topup_r[:,row]=nonlin_topup_r
        
        fmap_names=fmap_names+['nonlinear '+v]
        topup_names=topup_names+['nonlinear '+v]
        row+=1
        
    
    plot_list=[lin_fmap_r]
    for i in range(len(all_versions)):
        plot_list=plot_list+[all_nonlin_fmap_r[:,i]]
    
    topup_plot_list=[lin_topup_r]
    for i in range(len(all_versions)):
        topup_plot_list=topup_plot_list+[all_nonlin_topup_r[:,i]]
    
    
    name_list=['linear']+fmap_names
    topup_name_list=['linear']+topup_names
    
    
    if space=='indv':
        ax = fig.add_subplot(321)
    elif space=='mni':
        ax = fig.add_subplot(323)
    elif space=='mni_ofc':
        ax = fig.add_subplot(325)
        
    sns.violinplot(plot_list,names=name_list,
                inner="points", color=pal)
    sns.axlabel(space,"correlation to fieldmap\n[spearman's rho]")
    
    
        
    if space=='indv':
        ax = fig.add_subplot(322)
    elif space=='mni':
        ax = fig.add_subplot(324)
    elif space=='mni_ofc':
        ax = fig.add_subplot(326)
        
    sns.violinplot(topup_plot_list,names=topup_name_list,
                inner="points", color=pal)
    sns.axlabel(space,"correlation to topup\n[spearman's rho]")

plt.savefig(save_path+'group_meanepi_spearmancorr_violins.svg')
```
