### Plotting distributions

```python
sns.set_context='poster'
sns.set(style='darkgrid')
lin_color=(0.86, 0.37119999999999997, 0.34)
fmap_color=(0.9677975592919913, 0.44127456009157356, 0.5358103155058701)
topup_color=(0.75, 0.4, 0.5)
a_color=(0.8616090647292522, 0.536495730113334, 0.19548899031476086)
b_color=(0.8, 0.6151497514677574, 0.11111)
c_color= (0.46810256823426105, 0.6699492535792404, 0.1928958739904499)
d_color= (0.20125317221201128, 0.6907920815379025, 0.6)
e_color= (0.21044753832183283, 0.55, 0.7)
fmap_pal=sns.color_palette([#lin_color, 
                       #fmap_color, 
                       #topup_color, 
                       #a_color, 
                       b_color, 
                       c_color, 
                       d_color, 
                       e_color],
                      5)

topup_pal=sns.color_palette([#lin_color, 
                       fmap_color, 
                       #topup_color, 
                       #a_color, 
                       b_color, 
                       c_color, 
                       d_color, 
                       e_color],
                      6)
```

```python
fig=plt.figure(figsize=(20,10))

fmap_topup_r=np.load(save_path+'arrays/fmap_topup_field_corr.npy')

all_nonlin_fmap_r=np.zeros((len(fmap_topup_r),len(all_versions)))
all_nonlin_topup_r=np.zeros((len(fmap_topup_r),len(all_versions)))

fmap_names=[]
topup_names=[]


row=0
for v in all_versions:
    nonlin_fmap_r=np.load('/scr/kansas1/huntenburg/eval/version_'+v+'/figures/group/arrays/nonlin_fmap_field_corr.npy')
    nonlin_topup_r=np.load('/scr/kansas1/huntenburg/eval/version_'+v+'/figures/group/arrays/nonlin_topup_field_corr.npy')

    all_nonlin_fmap_r[:,row]=nonlin_fmap_r
    all_nonlin_topup_r[:,row]=nonlin_topup_r
    
    #fmap_names=fmap_names+['nonlinear '+v]
    fmap_names=['v1','v3','v5']
    #topup_names=topup_names+['nonlinear '+v]
    topup_names=['v1','v3','v5']
    row+=1
    

fmap_list=[]
for i in range(len(all_versions)):
    fmap_list=fmap_list+[all_nonlin_fmap_r[:,i]]
#fmap_list=[fmap_topup_r]+fmap_list


topup_list=[]
for i in range(len(all_versions)):
    topup_list=topup_list+[all_nonlin_topup_r[:,i]]
topup_list=[fmap_topup_r]+topup_list


#fmap_names=['topup']+fmap_names
fmap_names=fmap_names
topup_names=['fmap']+topup_names


fig.add_subplot(2,2,1)
sns.violinplot(fmap_list,names=fmap_names,
            inner="points", color=fmap_pal)
sns.axlabel('',"correlation to fieldmap field\n[spearman's rho]")

fig.add_subplot(2,2,2)
sns.violinplot(topup_list,names=topup_names,
            inner="points", color=topup_pal)
sns.axlabel('',"correlation to topup field\n[spearman's rho]")

plt.savefig(save_path+'group_field_spearmancorr_violins.svg')
```
