## Field correlations across subjects

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy as sp
import os
```

# CHECK BEFORE RUNNING

```python
all_versions = ['8','10','5','6', '11',
                '9','13','7','14','16',
                '2','17','15','18','4','19']
versions = ['version_8','version_10','version_5','version_6', 'version_11',
            'version_9','version_13','version_7','version_14','version_16',
            'version_12','version_17','version_15','version_18','version_4','version_19']
```

```python
subjects=[]
f = open('/scr/jessica2/Schaare/LEMON/done_freesurfer.txt','r')
for line in f:
    subjects.append(line.strip())
```

```python
correlations = ['nonlin_fmap', 'nonlin_topup','fmap_topup']
```

### Reading values into arrays

```python
for version in versions:
    
    file_path='/scr/kansas1/huntenburg/eval/'+version+'/%s/fieldcompare/field_correlation_indv.txt'
    save_path='/scr/kansas1/huntenburg/eval/'+version+'/figures/group/'

    if not os.path.isdir('/scr/kansas1/huntenburg/eval/'+version+'/figures/'):
        os.mkdir('/scr/kansas1/huntenburg/eval/'+version+'/figures/')



    for correlation in correlations:
        if correlation=='nonlin_fmap': 
            row=0
        if correlation=='nonlin_topup': 
            row=1
        if correlation=='fmap_topup': 
            row=2
                
        # initiate empty array and min/max values
        r_array=np.array([])
        r_min=[1, 'none']
        r_max=[0,'none']
            
        # loop over subjects, read in metrics file
        for subject in range(len(subjects)):
            path = file_path %(subjects[subject])
            stats_array=np.loadtxt(path, dtype='float', delimiter=' ')
            
            # define r values
            r_val=stats_array[row, 0]
            
            if isnan(r_val):
                print r_val, subjects[subject]
            
            # compare to min/max values and update if applicable
            if r_val < r_min[0]:
                r_min=[r_val,subjects[subject]]
            if r_val > r_max[0]:
                r_max=[r_val, subjects[subject]]
                
            # append values to arrays
            r_array = np.append(r_array,r_val)
            
        # save array of all subjects to file
        if not os.path.isdir(save_path):
            os.mkdir(save_path)
        
        if not os.path.isdir(save_path+'arrays/'):
            os.mkdir(save_path+'arrays/')
        np.save(save_path+'arrays/'+correlation+'_field_corr.npy', r_array)
    
```

### Plotting distributions

```python
sns.set_context='poster'
sns.set(style='darkgrid')
```

```python
fig=plt.figure(figsize=(15,10))

fmap_topup_r=np.load(save_path+'arrays/fmap_topup_field_corr.npy')

all_nonlin_fmap_r=np.zeros((len(fmap_topup_r),len(all_versions)))
all_nonlin_topup_r=np.zeros((len(fmap_topup_r),len(all_versions)))

fmap_names=[]
topup_names=[]


i=1
row=0
for v in all_versions:
    nonlin_fmap_r=np.load('/scr/kansas1/huntenburg/eval/version_'+v+'/figures/group/arrays/nonlin_fmap_field_corr.npy')
    nonlin_topup_r=np.load('/scr/kansas1/huntenburg/eval/version_'+v+'/figures/group/arrays/nonlin_topup_field_corr.npy')

    all_nonlin_fmap_r[:,row]=nonlin_fmap_r
    all_nonlin_topup_r[:,row]=nonlin_topup_r
    
    fmap_names=fmap_names+['v'+str(i)]
    #fmap_names=['v1','v2','v3','v4','v5']
    topup_names=topup_names+['v'+str(i)]
    #topup_names=['v1','v2','v3','v4','v5']
    row+=1
    i+=1
    

fmap_list=[]
for i in range(len(all_versions)):
    fmap_list=fmap_list+[all_nonlin_fmap_r[:,i]]
fmap_list=[fmap_topup_r]+fmap_list


topup_list=[]
for i in range(len(all_versions)):
    topup_list=topup_list+[all_nonlin_topup_r[:,i]]
topup_list=[fmap_topup_r]+topup_list


fmap_names=['topup']+fmap_names
#fmap_names=fmap_names
topup_names=['fmap']+topup_names


fig.add_subplot(2,1,1)
sns.violinplot(fmap_list,names=fmap_names,
            inner="points", color='husl')
sns.axlabel('',"correlation to fieldmap field\n[spearman's rho]")

fig.add_subplot(2,1,2)
sns.violinplot(topup_list,names=topup_names,
            inner="points", color='husl')
sns.axlabel('',"correlation to topup field\n[spearman's rho]")

plt.savefig('/scr/kansas1/huntenburg/eval/supplementary/real_fieldcorr.svg')
```
