## voxelwise correlation of simulated fields/images

```python
import nibabel as nb
import scipy as sp
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import os
```

# CHECK BEFORE RUNNING

```python
version='version_5'
dwell= '0007'
#contrast='field_corr'
#contrast='lin_truth_corr'
contrast='nonlin_truth_corr'
#contrast='fmap_truth_corr'
```

```python
filename = dwell+'_'+contrast
save_path='/scr/kansas1/huntenburg/eval/'+version+'/figures/simulated/'
if not os.path.isdir('/scr/kansas1/huntenburg/eval/'+version+'/figures/'):
    os.mkdir('/scr/kansas1/huntenburg/eval/'+version+'/figures/')
    
if not os.path.isdir(save_path):
    os.mkdir(save_path)

if contrast=='field_corr':
    image1='/scr/kansas1/huntenburg/eval/'+version+'/simulated/%s/fields/original_field_masked.nii.gz' %(dwell)
    image2='/scr/kansas1/huntenburg/eval/'+version+'/simulated/%s/fields/nonlin_field_masked.nii.gz' %(dwell)
    mask='/scr/kansas1/huntenburg/eval/possum/LEMON006/fmap_mask.nii.gz'
    
    image1_label='original field'
    image2_label='nonlinear field'

elif contrast=='lin_truth_corr':
    image1='/scr/kansas1/huntenburg/eval/'+version+'/simulated/%s/groundtruth/lin_coreg.nii.gz' %(dwell)
    image2='/scr/kansas1/huntenburg/eval/'+version+'/simulated/%s/corrected/lin_coreg.nii.gz' %(dwell)
    mask='/scr/kansas1/huntenburg/eval/possum/LEMON006/brain_mask.nii.gz'
    
    image1_label='undistorted'
    image2_label='distorted lin coreg'
    
elif contrast=='nonlin_truth_corr':
    image1='/scr/kansas1/huntenburg/eval/'+version+'/simulated/%s/groundtruth/lin_coreg.nii.gz' %(dwell)
    image2='/scr/kansas1/huntenburg/eval/'+version+'/simulated/%s/corrected/nonlin_coreg.nii.gz' %(dwell)
    mask='/scr/kansas1/huntenburg/eval/possum/LEMON006/brain_mask.nii.gz'
    
    image1_label='undistorted'
    image2_label='distorted nonlin coreg'
    
elif contrast=='fmap_truth_corr':
    image1='/scr/kansas1/huntenburg/eval/'+version+'/simulated/%s/groundtruth/lin_coreg.nii.gz' %(dwell)
    image2='/scr/kansas1/huntenburg/eval/'+version+'/simulated/%s/corrected/fmap_coreg.nii.gz' %(dwell)
    mask='/scr/kansas1/huntenburg/eval/possum/LEMON006/brain_mask.nii.gz'
    
    image1_label='undistorted'
    image2_label='distorted fmap coreg'
```

load data

```python
img1=nb.load(image1).get_data()
img2=nb.load(image2).get_data()
mask=nb.load(mask).get_data()
```

```python
print shape(img1)
print shape(img2)
print shape(mask)
```

reduce data to mask

```python
img1_masked=img1[mask>0]
img2_masked=img2[mask>0]
```

```python
print shape(img1.flatten())
print shape(img2.flatten())
print shape(img1_masked)
print shape(img2_masked)
```

plot regression

```python
sns.set_context(context='talk')

fig = plt.figure()
sns.jointplot(img1_masked, img2_masked, kind='kde', color=(0.46810256823426105, 0.6699492535792404, 0.1928958739904499), #'#070B19',
              xlim=(80,180), ylim=(80,180),
              stat_func=sp.stats.spearmanr)
sns.axlabel(image1_label,image2_label)

plt.savefig(save_path+filename+'_kde.svg')
```

```python
sns.set_context(context='talk')
fig = plt.figure()
sns.jointplot(img1_masked, img2_masked, kind='reg', color='#0B173B', stat_func=sp.stats.spearmanr)
sns.axlabel(image1_label,image2_label)

#plt.savefig(save_path+filename+'_reg.svg')
#plt.savefig(save_path+filename+'_reg.jpg')
```
