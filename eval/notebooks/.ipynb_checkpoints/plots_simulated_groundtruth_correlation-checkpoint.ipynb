## Correlation of lin/nonlin/fmap coregistration to groundtruth across different levels of distortion

importing libraries

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy as sp
import os
```

setting variables

```python
distortions=['0001', '0002', '0003', '0004', '0005', '0006', '0007', '0008', '0009', '0010']
coregs = ['lin', 'nonlin', 'fmap']
```

```python
version='version_1'
```

```python
corrected_txt='/scr/kansas1/huntenburg/eval/'+version+'/simulated/%s/corrected/correlation_groundtruth.txt'
groundtruth_txt='/scr/kansas1/huntenburg/eval/'+version+'/simulated/0001/groundtruth/correlation_lin_nonlin.txt'
save_path='/scr/kansas1/huntenburg/eval/'+version+'/figures/simulated/'
```

reading metrics and difference values into arrays, doing stats, saving to files

```python
# read groundtruth correlation (lin/nonlin) from file
groundtruth_array=np.loadtxt(groundtruth_txt, dtype='float', delimiter=' ')
groundtruth_nonlin=groundtruth_array[2]

for coreg in coregs:
    if coreg=='lin': 
        row=0
        r_nodistortion=1.0
    if coreg=='nonlin': 
        row=1
        r_nodistortion=groundtruth_nonlin
    if coreg=='fmap': 
        row=2
        r_nodistortion=1.0
            
    # initiate empty array and add value for no distortion
    r_array=np.array([])
    r_array=np.append(r_array, r_nodistortion)
    
    for distortion in distortions:
        path = corrected_txt %(distortion)
        stats_array=np.loadtxt(path, dtype='float', delimiter=' ')
        
        # get r value
        r_val=stats_array[row, 2]
            
        # append values to arrays
        r_array = np.append(r_array,r_val)
        
    # save array across distotions to file
    if not os.path.isdir(save_path):
        os.mkdir(save_path)
    
    if not os.path.isdir(save_path+'arrays/'):
        os.mkdir(save_path+'arrays/')
    np.save(save_path+'arrays/'+coreg+'_epi_r.npy', r_array)
```

```python
np.load(save_path+'arrays/'+'fmap'+'_epi_r.npy')
```

## Plotting correlations

```python
lin_r=np.load(save_path+'arrays/lin_epi_r.npy')
nonlin_r=np.load(save_path+'arrays/nonlin_epi_r.npy')
fmap_r=np.load(save_path+'arrays/fmap_epi_r.npy')
dwell=np.array([0,0.0001,0.0002,0.0003,0.0004,0.0005,0.0006,0.0007,0.0008,0.0009,0.001])

sns.set_context(context='talk')
fig = plt.figure()
ax = fig.add_subplot(111)
sns.pointplot(dwell, lin_r, color='ghostwhite',label='lin')
sns.pointplot(dwell, nonlin_r, color='lightblue', label='nonlin')
sns.pointplot(dwell, fmap_r, color='mediumseagreen', label='fmap')
sns.axlabel('echo spacing','R^2')


#plt.savefig(save_path+'r_violins.svg')
```
